```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import numpy as np

options = webdriver.ChromeOptions()

browser = webdriver.Chrome(options=options)

url = 'https://fundhot.com/article/list'
                                                                                                                                                                                                                                                                                                                                                                                                               
browser.get(url)


browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    
    
soup = BeautifulSoup(browser.page_source, 'html.parser') 
time = soup.find_all('div', {'class': 'forum-date'})
contents = soup.find_all('div', {'class': 'forum-title'})
link = soup.find_all('a', {'class': 'forum-row'})

time_record = []
title = []
links = []

for r_time in time:
    time_record.append(r_time.text.strip())
    
for review in contents:
    title.append(review.text.strip())
    
for link_record in link:
    links.append(link_record["href"])

all_data = list(zip(time_record,title,links))

df = pd.DataFrame(all_data, columns = ['日期','標題','內容連結'])

df.to_csv('review.csv',encoding='utf_8_sig')

df
```

```python
df1=df.drop(df.tail(1224).index)
df1
```

```python
import datetime
def get_week_of_month(year,month,day):
    begin = int(datetime.date(year, month, 1).strftime("%W"))
    end = int(datetime.date(year,month,day).strftime("%W"))

    return end - begin + 1
```

```python
week_id=[]
for time in df1["日期"]:
    y=int(time[0:4])
    m=int(time[5:7])
    d=int(time[8:10])
    week_index=get_week_of_month(y,m,d)
    week_id.append(time[0:4]+"-"+time[5:7]+"-"+str(week_index))
df1["週期ID"]=week_id
df1
```

```python
def find_content(link):
    
    s_list = []

    res2 = requests.get(link)
    soup2 = BeautifulSoup(res2.text, "html.parser")
    temp_content=soup2.find_all('div', attrs={"class": "postbody"})
    other=soup2.find_all('span', attrs={"style": "color: #0000FF"})
    for data in temp_content:
        a="".join(data.text)
        if "https" in a:
            continue
        if "寫:" in a:
            s_list.append(a.split(":")[-1])
            continue
        s_list.append(a)
    for word in other:
        s_list.append(word.text)
        
    return "".join(s_list)
```

```python
df1["內容"]=df1["內容連結"].apply(find_content)
```

```python
list3=[]
x=pd.Series(df1["內容"].index)
for i in x:
    if len(df1["內容"][i]) == 0:
        list3.append(i)
len(list3)
```

```python
df2=df1.drop(list3,axis=0)
df2.reset_index(inplace=True)
df2.drop("index",axis=1)
```
