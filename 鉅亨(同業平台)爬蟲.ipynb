```python
url="https://blog.anuefund.com/"
dates=[]
titles=[]
links=[]
times=1

for i in range(46):
    res = requests.get(url)
    soup=BeautifulSoup(res.text, "html.parser")
    articles=soup.select(".entry-title a")
    time = soup.select(".entry-date")
    times+=1
    next_url = "https://blog.anuefund.com/page/" + str(times) + "/"
    url=next_url
    for each_date in time:
        a=re.sub('[年月日]', '', each_date.text)
        dates.append(a.replace("  ","-").strip())        
    for each_title in articles:
        titles.append(each_title.text)
    for each_title in articles:
        links.append(each_title["href"] )

df = pd.DataFrame(
{
    '日期':dates,
    '標題':titles,
    '內容連結':links
})
df.to_csv('鉅亨(同業平台)標題資料.csv',encoding='utf_8_sig')
df
```

```python
def find_content(links):
    
    s_list = []

    res2 = requests.get(links)
    soup2 = BeautifulSoup(res2.text, "html.parser")
    temp_content1 = soup2.select(".entry-content li")
    temp_content2 = soup2.select(".entry-content p")
    for data in temp_content1:
        if data.text.__contains__("申購") or data.text.__contains__("資料來源"):
            continue            
        s_list.append(data.text.strip().replace("\ufeff\ufeff\ufeff\ufeff\ufeff",""))
    for data in temp_content2:
        if data.text.__contains__("申購") or data.text.__contains__("資料來源"):
            continue            
        s_list.append(data.text.strip().replace("\ufeff\ufeff\ufeff\ufeff\ufeff","")) 

    return "".join(s_list)
```

```python
df["內容"]=df["內容連結"].apply(find_content)
```

```python
import datetime

def get_week_of_month(year, month, day):
    begin = int(datetime.date(year, month, 1).strftime("%W"))
    end = int(datetime.date(year, month, day).strftime("%W"))

    return end - begin + 1
```

```python
week=[]
for time in df["日期"]:
    y=int(time.strftime("%Y"))
    m=int(time.strftime("%m"))
    d=int(time.strftime("%d"))
    week_id=get_week_of_month(y,m,d)
    week.append(time.strftime("%Y")+"-"+time.strftime("%m")+"-"+str(week_id))
df["週期ID"]=week
df
```
