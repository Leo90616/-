```python
url="https://www.ptt.cc/bbs/Fund/index.html"

titles=[]
links=[]
for times in range(25):
    res = requests.get(url)
    soup=BeautifulSoup(res.text, "html.parser")
    articles=soup.select(".title a")
    paging1=soup.select(".btn-group.btn-group-paging a")
    next_url = "https://www.ptt.cc" + paging1[1]["href"]
    url=next_url
    for each_title in articles:
        if each_title.text.__contains__("問卷") or each_title.text.__contains__("公告") or each_title.text.__contains__("閒聊"):
            continue
        titles.append(each_title.text)
        links.append("https://www.ptt.cc"+each_title["href"] )

df = pd.DataFrame(
{
    '標題':titles,
    '內容連結':links
})

df
```

```python
import datetime
from datetime import datetime

def find_date(link):
    res3 = requests.get(link)
    soup3 = BeautifulSoup(res3.text, "html.parser")
    date=soup3.select(".article-meta-value")
    time=date[-1].text.split()
    new_time=time[4]+"-"+time[1]+"-"+time[2]    
    return datetime.strptime(new_time,"%Y-%b-%d")
```

```python
df["日期"]=df["內容連結"].apply(find_date)
```

```python
import datetime
def get_week_of_month(year,month,day):
    begin = int(datetime.date(year, month, 1).strftime("%W"))
    end = int(datetime.date(year,month,day).strftime("%W"))

    return end - begin + 1
```

```python
week=[]
for time in df["日期"]:
    y=int(time.strftime("%Y"))
    m=int(time.strftime("%m"))
    d=int(time.strftime("%d"))
    week_id=get_week_of_month(y,m,d)
    week.append(time.strftime("%Y")+"-"+time.strftime("%m")+"-"+str(week_id))
df["週期ID"]=week
df
```

```python
df.index = df['日期']
df.sort_index(ascending=False)
```

```python
from lxml import etree


def find_content(link):
    s_list=[]


    res2 = requests.get(link)
    soup2 = BeautifulSoup(res2.text, "html.parser")
    temp_content = soup2.select(".push-content")
    selector = etree.HTML(res2.text)
    results = selector.xpath('//*[@id="main-content"]/text()') 
    rich_content=soup2.find_all('span', attrs={"class": "f6"})

    for text in results:
        a=text.replace("\n","").replace("--","").replace(" ","").strip()
        s_list.append(a)
        
    for x in rich_content:
        word=x.text.replace("\n","").replace(": ","")
        if "https" in word:
            continue
        if "※" in word:
            continue
        s_list.append(word)

    for data in temp_content:
        sentence=data.text.strip().replace(": ","").replace(" ","")
        if len(sentence)>4:
            if "https://" not in sentence:
                s_list.append(sentence)
    
    final="".join(s_list)

    return final
```

```python
df["內容"]=df["內容連結"].apply(find_content)
```

```python
df.to_csv("基金版(論壇)內文資料.csv",encoding='utf_8_sig' )
```
