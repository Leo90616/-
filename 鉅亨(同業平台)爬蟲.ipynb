```python
def find_content(links):
    
    s_list = []

    res2 = requests.get(links)
    soup2 = BeautifulSoup(res2.text, "html.parser")
    temp_content1 = soup2.select(".entry-content li")
    temp_content2 = soup2.select(".entry-content p")
    for data in temp_content1:
        if data.text.__contains__("ç”³è³¼") or data.text.__contains__("è³‡æ–™ä¾†æº"):
            continue            
        s_list.append(data.text.strip().replace("\ufeff\ufeff\ufeff\ufeff\ufeff",""))
    for data in temp_content2:
        if data.text.__contains__("ç”³è³¼") or data.text.__contains__("è³‡æ–™ä¾†æº"):
            continue            
        s_list.append(data.text.strip().replace("\ufeff\ufeff\ufeff\ufeff\ufeff","")) 

    return "".join(s_list)
```

```python
import datetime

def get_week_of_month(year, month, day):
    begin = int(datetime.date(year, month, 1).strftime("%W"))
    end = int(datetime.date(year, month, day).strftime("%W"))

    return end - begin + 1
```

```python
week=[]
for time in df["æ—¥æœŸ"]:
    y=int(time.strftime("%Y"))
    m=int(time.strftime("%m"))
    d=int(time.strftime("%d"))
    week_id=get_week_of_month(y,m,d)
    week.append(time.strftime("%Y")+"-"+time.strftime("%m")+"-"+str(week_id))
df["é€±æœŸID"]=week
df
```
