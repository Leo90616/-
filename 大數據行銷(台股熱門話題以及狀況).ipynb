```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import numpy as np

titles = []
inf = []
times=[]

options = webdriver.ChromeOptions()

browser = webdriver.Chrome(options=options)


for i in range(1,4):
    url = 'https://finance.ettoday.net/focus/175/'+str(i)

    browser.get(url)

    
    a=browser.find_element_by_class_name("btn.btn_close")
    if a==True:
        a.click()

    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")



    soup = BeautifulSoup(browser.page_source, 'html.parser') 
    contents = soup.select(".part_pictxt_3 a")
    dates= soup.select(".date")
    flink= soup.find_all('a', {'class': 'piece clearfix'})


    for title in contents:
        titles.append(title["title"].replace("\u3000",""))

    for i in flink:
        inf.append(i.text.strip().split("\n")[1])
            
    for time in dates:
        if "小時" in time.text:
            times.append("2021-06-25")
            continue
        times.append(time.text.split(" ")[0])

all_data = list(zip(times,titles,inf))

df = pd.DataFrame(all_data, columns = ['日期','標題','大概內容'])

df
```

```python
def tool(i):
    
    if "航" in i:
        return "航運"
    if "運" in i:
        return "航運"
    if "聯亞" in i:
        return "疫苗"
    if "高端" in i:
        return "疫苗"
    if "台積電" in i:
        return "科技"
    if "聯發科" in i:
        return "科技"
    return "無"
    
    
```

```python
df["分類"]=df["大概內容"].apply(tool)
```

```python
df
```

```python
df=df[~df['分類'].isin(["無"])]
df.reset_index(inplace=True)
```

```python
df1=df.drop(["index"],axis=1)
```

```python
df1
```

```python
list2=[]
for i in df1["大概內容"]:
    if "漲" in i:
        list2.append("1")
        continue
    if "跌" in i:
        list2.append("0")
        continue
    if "弱" in i:
        list2.append("0")
        continue
    if "下調" in i:
        list2.append("0")
        continue
    list2.append(None)
df1["狀況"]=list2
```

```python
df2=df1.dropna()
```

```python
freq = df2['分類'].value_counts() 
result=pd.Series(freq.sort_index())
result
```

```python
def freq_tool(x):
    freq = df1[df1['狀況']==x]["分類"].value_counts() 
    result=pd.Series(freq.sort_index())
    return result
```

```python
list3=list(freq_tool("0"))
list4=list(freq_tool("1"))
list4
```

```python
list3
```

```python
import plotly as py
import plotly.graph_objs as go
pyplt = py.offline.plot

# Stacked Bar Chart
trace_0 = go.Bar(
    x = ['疫苗股', '科技股','航運股'],
    y = list3,
    name = '下跌新聞篇數',
    marker=dict(
        color='green')

)

trace_1 = go.Bar(
    x = ['疫苗股', '科技股','航運股'],
    y = list4,
    name = '上漲新聞篇數',
    marker=dict(
        color='red')
)


trace = [trace_0, trace_1]
layout = go.Layout(
    title = '近期三大熱門台股話題聲量以及狀況',
    barmode='stack',
    legend_font_size=20,
    font_size=15,
    font_family="Arial"
)

fig = go.Figure(data = trace, layout = layout)
# pyplt(fig, filename='tmp_1.html')
fig
```
